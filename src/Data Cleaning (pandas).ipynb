Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Aterbruket\Documents\Customer Call List.xlsx")

df
```

```python
df = df.drop_duplicates()

df
```

```python
df.drop(columns = "Not_Useful_Column")
```

```python
df.drop(columns = 'Not_Useful_Column', inplace = True)
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
```

```python
df
```

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.rstrip("_")
```

```python
df
```

```python
df['Last_Name'].str.strip("._/123")
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("._/123")
```

```python
df
```

```python
df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
```

```python
df
```

```python
df["Address"].str.split(',', n = 1)
```

```python
df["Address"].str.split(',', n = 1, expand = True)
```

```python
df['Address'].str.split(',', n = 2, expand = True)
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', n = 2, expand = True)
```

```python
df
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
```

```python
df
```

```python
df = df.replace('N/a', '')

df
```

```python
df = df.fillna('')
```

```python
df
```

```python
#Picking Customers that say it is okay to call
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('', 'N')
```

```python
df
```

```python
#Droping custormers without a phone number
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df

# Another way to drop null values

#df.dropna(subset = "Phone_Number", inplace = True)
```
